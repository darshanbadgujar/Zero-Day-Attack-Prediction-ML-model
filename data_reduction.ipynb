```python
import pandas as pd
import numpy as np
```

```python
#first dataset
data = pd.read_csv("F:/Internship/Dataset/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")
```

```python
data.head(10)
```

```python
def entropy(s):
    values, counts = np.unique(s, return_counts=True)
    probabilities = counts / len(s)
    entropy = -np.sum(probabilities * np.log2(probabilities))
    return entropy
```

```python
df = pd.DataFrame(data)
```

```python
#Calculate entropy
for i in df.columns:
    print(i,"   ",entropy(df[i]))
```

```python
#dropped destination port
df = df.drop([' Destination Port'], axis=1)
```

```python
#dropped columns with entropy 0
cols_with_entropy_0 = []
for i in list(df.columns):
    if entropy(df[i]) == 0:
        cols_with_entropy_0.append(i)
df = df.drop(cols_with_entropy_0, axis = 1)
```

```python
#Replaced name BENIGN - 0, DDoS - 1
df[' Label'].replace({'BENIGN': 0, 'DDoS': 1}, inplace=True)
```

```python
print(df)
```

```python
#Export new csv
csv_file_path = 'F:/Internship/Dataset/NEW DATASET/Friday_afternoon_DDoS.csv'
df.to_csv(csv_file_path, index=False)
```
