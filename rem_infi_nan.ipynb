```python
import pandas as pd
import numpy as np
```

```python
#Import Data
data1 = pd.read_csv("F:/Internship/Dataset/NEW DATASET/Friday_afternoon_DDoS.csv")
```

```python
df = pd.DataFrame(data1)
```

```python
special_rows = df[df.applymap(np.isnan).any(axis=1) | df.applymap(np.isinf).any(axis=1)]
```

```python
print(special_rows)
```

```python
df = df.drop(special_rows.index)
```

```python
df
```

```python
#Export new csv
csv_file_path = 'F:/Internship/Dataset/Cleaner Data/Fri_Afternoon_DDoS.csv'
df.to_csv(csv_file_path, index=False)
```
